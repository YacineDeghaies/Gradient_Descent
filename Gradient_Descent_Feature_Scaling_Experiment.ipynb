```python
from sklearn.linear_model import SGDRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.array([[1, 2],
              [2, 4],
              [3, 6],
              [4, 8],
              [5, 100]]
)

y = np.array([5, 7, 9, 11, 13])
```

```python
def cost_function(X, y, model, num_iter=1000):
  sum_cost = []
  for _ in range(num_iter):
    model.fit(X, y)
    y_pred = model.predict(X)
    cost = mean_squared_error(y_pred, y)
    sum_cost.append(cost)

  return sum_cost
```

```python
alpha = 0.01
model = SGDRegressor(learning_rate="constant", eta0=alpha, tol=None, warm_start=True)
```

```python
#track the loss
loss_curve = cost_function(X, y, model)
plt.plot(loss_curve)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.show()
```

```python
#Normalize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

#Re-initialize the model after scaling the data

model_scaled = SGDRegressor(learning_rate="constant", eta0=alpha, tol=None, warm_start=True)

# Track the loss for the scaled data
loss_curve_scaled = cost_function(X_scaled, y, model)

#Plot the Loss values after feature scaling
plt.plot(loss_curve_scaled)
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.show()
```

# Effect of Feature Scaling on Gradient Descent
Upon plotting the loss function values of the first 10 iterations, it becomes evident that employing feature scaling accelerates the convergence of our loss function much more faster that without feature Scaling.
